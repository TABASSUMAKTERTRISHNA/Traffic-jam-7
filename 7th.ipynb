```python
import matplotlib.pyplot as plt
```

```python
times=['12 am', '1 am', '2 am', '3 am', '4 am', '5 am', '6 am', '7 am', '8 am', '9 am', '10 am', '11 am', '12 pm', '1 pm', '2 pm', '3 pm', '4 pm', '5 pm', '6 pm', '7 pm', '8 pm', '9 pm', '10 pm', '11 pm']
```

```python
jam_level=[1,2,1,3,2,3,4,5,4,3,2,1,4,5,6,5,4,3,2,1,1,2,3,4]
```

```python
time=input("enter a time : ")
```

```python
if time in times:
    index=times.index(time)
    plt.plot(times[index:index+12],jam_level[index:index+12],color='red')
    plt.xlabel('time of day')
    plt.ylabel('traffic jam level')
    plt.title('trafiic jam at different times in a 12 hour period')
    plt.show()
else:
    print('invalid time')
```
